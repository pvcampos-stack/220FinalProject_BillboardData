```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("hhi_country.csv")
print(df.head())

top10 = df.sort_values("HHI", ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top10["chart"], top10["HHI"])
plt.xlabel("HHI")
plt.title("Top 10 Charts by HHI")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gini_country.csv")

top10 = df.sort_values("Gini", ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top10["chart"], top10["Gini"])
plt.xlabel("Gini Coefficient")
plt.title("Top 10 Charts by Streaming Inequality")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cr3_country.csv")

top10 = df.sort_values("CR3", ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top10["chart"], top10["CR3"])
plt.xlabel("CR3 (Top 3 Songs Share)")
plt.title("Top 10 Charts by Top-3 Song Concentration")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
df = pd.read_excel("concentration_results.xlsx")

top = df.sort_values("HHI", ascending=False).head(8)

top.set_index("chart")[["HHI","Gini"]].plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Comparison of HHI and Gini Across Charts")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
global_songs = pd.read_sql("""
SELECT song_name, artist, rank
FROM BillboardCharts
WHERE chart = 'Worldwide Top 200 Songs'
AND rank <= 10
ORDER BY rank
""", conn)

global_songs
```

```python
spotify_matches = pd.read_sql("""
SELECT DISTINCT song_name, artist
FROM SpotifyWeeklyStreamingCharts
WHERE artist LIKE '%Bad Bunny%'
ORDER BY song_name
""", conn)

spotify_matches
```

```python
song = global_songs.iloc[0]["song_name"]
artist = global_songs.iloc[0]["artist"]

song_streams = pd.read_sql(f"""
SELECT country, SUM(Total) AS streams
FROM SpotifyWeeklyStreamingCharts
WHERE country <> 'Global'
  AND LOWER(song_name) = LOWER("{song}")
  AND LOWER(artist) = LOWER("{artist}")
GROUP BY country
ORDER BY streams DESC
""", conn)

song_streams
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"

fig = px.choropleth(
    song_streams,
    locations="country",
    locationmode="country names",
    color="streams",
    hover_name="country",
    hover_data={"streams": True},
    title=f"Geographic Distribution of Spotify Streams for '{song}'",
    color_continuous_scale="Viridis"
)

fig.show()
```

```python
import requests
import pandas as pd
from io import StringIO

url = "https://www.worldometers.info/world-population/population-by-country/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
html = response.text

pop_tables = pd.read_html(StringIO(html))
population_df = pop_tables[0]

population_df.head()
```

```python
population_df.columns
```

```python
population_df = population_df.iloc[:, [1, 2]]
population_df.columns = ["country", "population"]
population_df.head()
```

```python
import sqlite3

conn = sqlite3.connect("data.db")
```

```python
song = "DTMF"
artist = "Bad Bunny"

spotify_song = pd.read_sql(f"""
SELECT country, SUM(Total) AS streams
FROM SpotifyWeeklyStreamingCharts
WHERE country <> 'Global'
AND LOWER(song_name) = LOWER('{song}')
AND LOWER(artist) = LOWER('{artist}')
GROUP BY country
ORDER BY streams DESC
""", conn)

spotify_song
```

```python
merged_df = spotify_song.merge(population_df, on="country", how="left")
merged_df.head()
```

```python
merged_df["streams_per_million"] = (merged_df["streams"] / merged_df["population"]) * 1_000_000
merged_df.head()
```

```python
import plotly.express as px

fig = px.choropleth(
    merged_df,
    locations="country",
    locationmode="country names",
    color="streams_per_million",
    hover_name="country",
    hover_data={
        "streams": True,
        "population": True,
        "streams_per_million": True
    },
    title=f"Spotify Streams per Million People for '{song}'",
    color_continuous_scale="Viridis"
)

fig.show()
```

```python
fig.write_html("dtmf_normalized_map.html")
```
